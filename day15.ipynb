```python
### Part 1 ###

def get_input_numbers():
    with open('day15_input.txt') as f:
        return [int(x) for x in f.read().split(',')]

input_numbers = get_input_numbers()

number_to_round_last_spoken = {}


# We're going to do something awkward.
# We're going to lock-in the last_spoken amounts for all but the last one
# Reason being, is that on the round after the last number, we want to know when that last number was spoken.
# If we've already "locked-in" the last number, we have no idea when (or if) it was last said.
# Thus, saving the lock-in for when we're 2 rounds ahead will be a better idea.

def init_input_and_return_number_to_lock_in():
    for index in range(0, len(input_numbers) - 1):
        current_num_in_list = input_numbers[index]
        print(f"Current round: {index + 1}, Number: {current_num_in_list}")
        number_to_round_last_spoken[current_num_in_list] = index + 1

    return input_numbers[-1]

def get_number_at_round(round: int):
    number_to_lock_in = init_input_and_return_number_to_lock_in()
    current_round = len(input_numbers) + 1

    while(current_round < round):
        # At this point:
        # number_to_lock_in: number that was said in the round previous, but whose round number 
        #   has yet to be updated in number_to_round_last_spoken. We'll lock this in at the end.
        if number_to_lock_in in number_to_round_last_spoken:
            number_said_this_round = (current_round - 1) - number_to_round_last_spoken[number_to_lock_in]
        else:
            number_said_this_round = 0
        print(f"Current round: {current_round}, Number: {number_said_this_round}")
        # Let's lock in the previous number's round
        number_to_round_last_spoken[number_to_lock_in] = (current_round - 1)
        number_to_lock_in = number_said_this_round
        current_round = current_round + 1


get_number_at_round(2021)
```
