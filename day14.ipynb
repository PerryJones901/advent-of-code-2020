```python
### Part 1 ###

def get_input_lines():
    with open('day14_input.txt') as f:
        return [x for x in f.read().splitlines()]

input_lines = get_input_lines()
mask = 'X' * 36
memory_bank = {}

def dec_num_to_binary_len_36(dec_num: int) -> str:
    return format(dec_num, 'b').rjust(36, '0')

def handle_mask_line(mask_line: str) -> str:
    return mask_line.split(' = ')[1]

def handle_mem_line(mem_line: str) -> int:
    # split_str[0] = 'mem[123]'
    # split_str[1] = '456'
    #
    # so split_str[0][4:-1] = 123

    split_str = mem_line.split(' = ')
    pos = int(split_str[0][4:-1])
    val = int(split_str[1])
    memory_bank[pos] = apply_mask(val)

def apply_mask(input: int) -> int:
    # STEP 1: convert X's to 1's and apply AND (forces 0s in mask to be written)
    # STEP 2: convert X's to 0's and apply OR (forces 1s in mask to be written)

    and_submask = int(mask.replace('X', '1'), 2)
    result_after_and_submask = and_submask & input

    or_submask = int(mask.replace('X', '0'), 2)
    return result_after_and_submask | or_submask

for line in input_lines:
    if(line[0:4] == 'mask'):
        mask = handle_mask_line(line)
    elif(line[0:3] == 'mem'):
        handle_mem_line(line)
    else:
        print('Error: line did not start with mask nor mem')
        break
# print(memory_bank)

sum = 0
for pos in memory_bank:
    sum += memory_bank[pos]
print(sum)
```

```python
### Part 2 ##

def get_input_lines():
    with open('day14_input.txt') as f:
        return [x for x in f.read().splitlines()]

input_lines = get_input_lines()
mask = '0' * 36
memory_bank = {}

def dec_num_to_binary_len_36(dec_num: int) -> str:
    return format(dec_num, 'b').rjust(36, '0')

def handle_mask_line(mask_line: str) -> str:
    return mask_line.split(' = ')[1]

def handle_mem_line(mem_line: str) -> int:
    # split_str[0] = 'mem[123]'
    # split_str[1] = '456'
    #
    # so split_str[0][4:-1] = 123

    split_str = mem_line.split(' = ')
    pos = int(split_str[0][4:-1])
    val = int(split_str[1])
    pos_after_mask = apply_mask(pos)
    set_val_in_pos(val, pos_after_mask)

def apply_mask(position: int) -> str:
    # '0 < 1 < X' - want to take highest in each index
    pos_as_binary_word = dec_num_to_binary_len_36(position)
    result = ''
    for i in range (0, 36):
        if (pos_as_binary_word[i] == 'X' or mask[i] == 'X'):
            result += 'X'
        elif (pos_as_binary_word[i] == '1' or mask[i] == '1'):
            result += '1'
        else:
            result += '0'
    return result

def set_val_in_pos(value: int, pos: str):
    first_x_to_zero = pos.replace('X', '0', 1)
    first_x_to_one = pos.replace('X', '1', 1)
    
    if (pos == first_x_to_zero):
        # no need to check first_x_to_one, as pos will equal that if equal to first_x_to_zero
        pos_to_int = int(pos, 2)
        memory_bank[pos_to_int] = value
    else:
        set_val_in_pos(value, first_x_to_zero)
        set_val_in_pos(value, first_x_to_one)

for line in input_lines:
    if(line[0:4] == 'mask'):
        mask = handle_mask_line(line)
    elif(line[0:3] == 'mem'):
        handle_mem_line(line)
    else:
        print('Error: line did not start with mask nor mem')
        break
# print(memory_bank)

sum = 0
for pos in memory_bank:
    sum += memory_bank[pos]
print(sum)
```
