```python
### Part 1 ###

REQUIRED_FIELDS = [
    "byr",
    "iyr",
    "eyr",
    "hgt",
    "hcl",
    "ecl",
    "pid",
    # "cid",
]

def getPassportsFromInput():
    with open('day4_input.txt') as f:
        return [x.replace('\n', ' ') for x in str.split(f.read(), '\n\n')]

def passportDict(passport_str: str):
    passport_dict = {}
    key_val_pairs = str.split(passport_str, ' ')
    for key_val_pair in key_val_pairs:
        key, value = str.split(key_val_pair, ':')
        passport_dict[key] = value

    return passport_dict

def passportHasAllRequiredFields(passport_dict: dict):
    for required_field in REQUIRED_FIELDS:
        if required_field not in passport_dict:
            return False
    return True

valid_passports = 0

for passport_str in getPassportsFromInput():
    passport_dict = passportDict(passport_str)
    if (passportHasAllRequiredFields(passport_dict)):
        valid_passports = valid_passports + 1

print(valid_passports)
```

```python
### Part 2 ###

import re

REQUIRED_FIELDS = ["byr", "iyr", "eyr", "hgt", "hcl", "ecl", "pid",
    # "cid",
]

ECL_COLOURS = ["amb", "blu", "brn", "gry", "grn", "hzl", "oth"]

def is_str_numeric_and_between(value: str, min: int, max: int):
    return value.isnumeric() and (min <= int(value) <= max)

def byr_validation(value: str):
    return len(value) == 4 and is_str_numeric_and_between(value, 1920, 2002)

def iyr_validation(value: str):
    return len(value) == 4 and is_str_numeric_and_between(value, 2010, 2020)

def eyr_validation(value: str):
    return len(value) == 4 and is_str_numeric_and_between(value, 2020, 2030)

def hgt_validation(value: str):
    if(value.endswith('cm')):
        str_without_cm = value[:-2]
        return is_str_numeric_and_between(str_without_cm, 150, 193)
    if(value.endswith('in')):
        str_without_in = value[:-2]
        return is_str_numeric_and_between(str_without_in, 59, 76)
    return False

def hcl_validation(value: str):
    return re.search('^#[a-f0-9]{6}$', value) is not None

def ecl_validation(value: str):
    return value in ECL_COLOURS

def pid_validation(value: str):
    return len(value) == 9 and value.isnumeric()

REQUIRED_FIELDS_WITH_VALIDATIONS = [
    ["byr", byr_validation],
    ["iyr", iyr_validation],
    ["eyr", eyr_validation],
    ["hgt", hgt_validation],
    ["hcl", hcl_validation],
    ["ecl", ecl_validation],
    ["pid", pid_validation],
]

def getPassportsFromInput():
    with open('day4_input.txt') as f:
        return [x.replace('\n', ' ') for x in str.split(f.read(), '\n\n')]

def passportDict(passport_str: str):
    passport_dict = {}
    key_val_pairs = str.split(passport_str, ' ')
    for key_val_pair in key_val_pairs:
        key, value = str.split(key_val_pair, ':')
        passport_dict[key] = value

    return passport_dict

def passportPassesValidation(passport_dict: dict):
    for required_field_with_validation in REQUIRED_FIELDS_WITH_VALIDATIONS:
        key = required_field_with_validation[0]
        validation = required_field_with_validation[1]
        if required_field_with_validation[0] not in passport_dict:
            return False
        if not validation(passport_dict[key]):
            return False
    return True

valid_passports = 0

for passport_str in getPassportsFromInput():
    passport_dict = passportDict(passport_str)
    if (passportPassesValidation(passport_dict)):
        valid_passports = valid_passports + 1

print(valid_passports)
```
